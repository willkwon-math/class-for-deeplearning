```
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init

num_data = 1000
num_epoch = 10000


## generating the data

num_data = 1000
num_epoch = 500
loss_arr = []

x=init.uniform_(torch.Tensor(num_data,1),-10,10)
noise = init.normal_(torch.FloatTensor(num_data,1),std=1)
y = (x**2)+3
y_noise = y + noise

# model

model = nn.Sequential(
 nn.Linear(1,6),
 nn.ReLU(),
 nn.Linear(6,10), 
 nn.ReLU(),
 nn.Linear(10,6),
 nn.ReLU(),
 nn.Linear(6,1),
)
loss_func = nn.L1Loss()
optimizer = optim.SGD(model.parameters(),lr=0.01)

#learning

label = y_noise
for i in range(num_epoch):
  optimizer.zero_grad()
  output = model(x)

  loss = loss_func(output,label)
  loss.backward()
  optimizer.step()

  if i % 10 ==0:
    print(loss.data)
  loss_arr.append(loss)
  
param_list = list(model.parameters()) 
```

```
import matplotlib.pyplot as plt

plt.plot(loss_arr)
plt.show()
```
